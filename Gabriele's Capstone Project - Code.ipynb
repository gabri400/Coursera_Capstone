We transform and visualize the results of the request in a json file.

```python
dromerestaurants = data.json()["response"]
dromerestaurants
```

We create a dataframe with the relevant informations we need to make the analysis and then we visualize it.

```python
items = dromerestaurants["groups"][0]["items"]

dfromerestaurants_raw = []
for item in items:
    venue = item["venue"]
    categories, uid, name, location = venue["categories"], venue["id"], venue["name"], venue["location"]
    assert len(categories) == 1
    shortname = categories[0]["shortName"]
    address = location["formattedAddress"]
    if not "postalCode" in location:
        continue
    postalcode = location["postalCode"]
    lat = location["lat"]
    lng = location["lng"]
    datarow = (uid, name, shortname, address, postalcode, lat, lng)
    dfromerestaurants_raw.append(datarow)
dfromerestaurants = pd.DataFrame(dfromerestaurants_raw, columns=["Restaurant_ID", "Restaurant_name", "Restaurant_shortname", "Restaurant_address", "Restaurant_postalcode", "Latitude", "Longitude"])
dfromerestaurants
```

We see the results in a map, using Folium, observing what areas are better to open a new restaurant, essentially watching where there is greater concentration of restaurants.

```python
rome_center = dromerestaurants["geocode"]["center"]


from folium import plugins

rome_map = folium.Map(location=[rome_center["lat"], rome_center["lng"]], zoom_start=14)

def add_markers(dfromerestaurants):
    for (j, row) in dfromerestaurants.iterrows():
        label = folium.Popup(row["Restaurant_name"], parse_html=True)
        folium.CircleMarker(
            [row["Latitude"], row["Longitude"]],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(rome_map)

add_markers(dfromerestaurants)
hm_data = dfromerestaurants[["Latitude", "Longitude"]].as_matrix().tolist()
rome_map.add_child(plugins.HeatMap(hm_data))

rome_map
```

In the end we can say that there are 4 places where it is advisable to open a restaurant in Rome, and they are highlighted in the map we have created.
